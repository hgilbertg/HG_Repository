```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(5)
```

```python
#Pregunta 1
sns.histplot(x = df['diastol'],
             bins=15,
           ).set_title('Histplot de diastol')

plt.show()
```

```python
#Pregunta 2
sns.lmplot(x='sistol', 
           y='peso', 
           data = df,
           hue='sexo',
           fit_reg = False, 
          )
```

```python
#Pregunta 3
sns.boxplot(x = df['sexo'],
            y = df['cintura'],
           ).set_title('Box plot de sexo - cintura')
plt.show()
```

```python
#Pregunta 4
sns.lmplot(x='peso', 
           y='cintura', 
           data = df,
           fit_reg = False,
          )
```
